```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Distribution Plots

```python
sns.distplot(tips['total_bill'],kde=False,bins=40)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)#kind=scatter(by default)/regression/hex/kde
```

```python
sns.pairplot(tips,hue='sex',palette='husl')
```

```python
sns.pairplot(tips)
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

# Categorical Plots

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar',hue='sex')
```

# Matrix Plot

```python
tips = sns.load_dataset('tips')
flight= sns.load_dataset('flights')
flight.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
fp=flight.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,linecolor='white',linewidth=1)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)

```

# Grid

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.barplot)
g.map_lower(sns.kdeplot)
```
