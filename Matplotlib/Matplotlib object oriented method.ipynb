```python
from matplotlib import pyplot as plt
```

```python
import numpy as np   
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,label='square')
axes.plot(x,x**3,label='cube')
axes.set_xlabel('xlabel')
axes.set_ylabel('ylabel')
axes.set_title('title')
axes.legend(loc=0)
```

```python
fig = plt.figure()
axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.2,0.5,0.4,0.3])
axes1.set_title('Bigger plot')

axes1.plot(x,y,'r')
axes2.set_title('Smaller plot')
axes2.plot(y,x,'g')
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,4),dpi=75)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture',dpi=200)
```

# Plot customization

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='g',lw=3,ls='--')#RGB hex codes can also be passed in color.For linewidth lw also work
ax.set_xlim([0,10])
```
